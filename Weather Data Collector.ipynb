```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = "https://www.timeanddate.com/weather/india/mumbai"

response = requests.get(url)
response
```

```python
city = soup.find("h1").text.strip()
```

```python
temperature = soup.find("div", class_="h2")
temperature = temperature.text.strip() if temperature else "N/A"
```

```python
condition = soup.find("div", class_="h2")
condition = condition.find_next("p").text.strip() if condition else "N/A"
```

```python
details = soup.find_all("tr")
weather_data = {}
```

```python
for detail in details:
    th = detail.find("th")
    td = detail.find("td")
    if th and td: 
        label = th.text.strip()
        value = td.text.strip()
        weather_data[label] = value
```

```python
df = pd.DataFrame(list(weather_data.items()), columns=["Parameter", "Value"])
```

```python
print(f"üìç City: {city}")
print(f"üå° Temperature: {temperature}")
print(f"‚õÖ Condition: {condition}\n")
print("üìä Additional Weather Details:\n")
display(df)
```

```python
filename = "mumbai_weather.csv"
df.to_csv(filename, index=False)
```
